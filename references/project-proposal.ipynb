### **Research Question**
Die zentrale Forschungsfrage lautet:  
**Welche Faktoren beeinflussen die Liefertermintreue bei produktive Materialien, und welche Erkenntnisse können aus der Analyse der verfügbaren Daten gewonnen werden, um Abweichungen zwischen geplanten und tatsächlichen Lieferterminen besser zu verstehen?**

Zusätzlich sollen folgende Fragestellungen explorativ untersucht werden:

   - Inwiefern beeinflussen Merkmale wie Lieferantentyp, Standort und Produktkategorie die Termintreue?

   - Welche zeitlichen Trends oder Muster lassen sich in Abweichungen von Liefermengen und Terminen identifizieren?

   - Wie zuverlässig sind die verfügbaren Daten für prädiktive Analysen, und welche statistischen Modelle könnten zur Vorhersage von Lieferabweichungen herangezogen werden?

### **Hypothesen**
Zur Beantwortung der Forschungsfrage und der explorativen Fragestellungen werden folgende Hypothesen formuliert:
1. **Hypothese 1: Einfluss von Lieferantenstandorten**
Die Liefertermintreue variiert signifikant in Abhängigkeit vom geografischen Standort der Lieferanten.

- Begründung: Lieferantenstandorte beeinflussen Faktoren wie Transportzeiten, Zollabwicklungen und logistische Komplexität, die sich direkt auf die Einhaltung von Lieferterminen auswirken können.

2. **Hypothese 2: Einfluss von Produktkategorien**
Die Produktkategorie beeinflusst die Häufigkeit und das Ausmaß von Abweichungen zwischen geplanten und tatsächlichen Lieferterminen.

- Begründung: Kauf- und Normteile mit komplexeren Produktionsanforderungen oder solchen, die aus spezifischen Rohstoffen gefertigt werden, könnten anfälliger für Lieferverzögerungen sein.

3. **Hypothese 3: Identifikation zeitlicher Muster**
Es bestehen signifikante zeitliche Muster in den Lieferabweichungen, die auf saisonale Schwankungen, Kapazitätsengpässe oder andere zeitliche Faktoren zurückzuführen sind.

- Begründung: Lieferketten sind häufig durch saisonale und zyklische Schwankungen geprägt, wie erhöhte Nachfrage zu Jahreszeiten oder Produktionsausfälle aufgrund von Feiertagen oder anderen Störungen.

### **Ziel der Untersuchung**
Das Projekt verfolgt das Ziel, die Einflussfaktoren auf die Liefertermintreue bei Kauf- und Normteilen (KON-Teilen) systematisch zu identifizieren und zu analysieren. Besonderes Augenmerk liegt auf den Auswirkungen geografischer Merkmale der Lieferantenstandorte, produktbezogener Kategorien und zeitlicher Muster.    •	Methode: Vergleich der Termintreue zwischen verschiedenen Lieferantenstandorten (z. B. national vs. international) mithilfe von Chi-Quadrat-Tests oder ANOVA.

Analyse, ob bestimmte Produktkategorien anfälliger für Lieferverzögerungen oder -abweichungen sind.

H3: Identifikation zeitlicher Muster in der Termintreue
Untersuchung, ob saisonale oder monatliche Trends in den Lieferabweichungen erkennbar sind.
   
   •	Methode: Zeitreihenanalysen zur Identifikation und Prognose von Trends in der Liefertermintreue.

### **Beschreibung der Zielvariable**

Die primäre Zielvariable in dieser Analyse ist **„Liefertermintreue“**.  
Diese Variable misst, ob eine Lieferung innerhalb des definierten Zeitrahmens im Vergleich zum geplanten Lieferdatum erfolgt ist. Sie wird wie folgt kategorisiert:  
- **Pünktlich**: Lieferung erfolgte innerhalb des akzeptablen Zeitfensters.  
- **Verspätet**: Lieferung erfolgte nach dem geplanten Datum.  
- **Früh**: Lieferung erfolgte vor dem geplanten Datum.  

Für bestimmte Analysen kann die Variable auch in ein binäres Format umgewandelt werden:  
- **1 (Pünktlich)**: Lieferung erfolgte planmäßig.  
- **0 (Nicht pünktlich)**: Lieferung erfolgte entweder früh oder verspätet.
