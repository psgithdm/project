```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import openpyxl
import joblib

# Laden des DataFrames
import_path = "../data/pkl/cleaned_anomalies_liefertreue_2024_df.pkl"
df = joblib.load(import_path)

# Überprüfen
print(df.head())

# Übersicht über die Daten
print("Erste fünf Zeilen der Daten:")
print(df.head())

# Visualisierung der Zielvariable
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Liefertreue (Ja/Nein)')
plt.title('Verteilung der Liefertreue')
plt.xlabel('Liefertreue (Ja/Nein)')
plt.ylabel('Anzahl')
plt.show()
```
