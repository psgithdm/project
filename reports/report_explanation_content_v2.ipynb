## **Einführung**

### **Begriffe und Definitionen**
- Liefertreue: Prozentualer Anteil der Lieferungen, die pünktlich und vollständig erfolgt sind.
- OTD-Rate (On-Time Delivery): Anteil der Lieferungen, die termingerecht geliefert wurden.
- OTIF-Rate (On-Time In-Full): Anteil der Lieferungen, die termingerecht und in der richtigen Menge geliefert wurden.
- Mengenabweichung (Ist vs. Soll): Differenz zwischen der gelieferten und der angeforderten Menge.
